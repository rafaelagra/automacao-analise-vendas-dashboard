```python
# 1- IMPORTANDO BIBLIOTECAS

# pandas: para carregar, tratar e analisar dados
# matplotlib.pyplot: para gerar gráficos simples e personalizáveis
# seaborn: para criar gráficos mais bonitos e informativos

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2- IMPORTAR O ARQUIVO EXCEL DO COMPUTADOR

from google.colab import files

# abre um botão para você selecionar o arquivo no seu computador
uploaded = files.upload()

# lê o primeiro arquivo excel enviado
import io
df = pd.read_excel(io.BytesIO(next(iter(uploaded.values()))))
```

```python
# 3- PRÉ-VISUALIZANDO DADOS

print("Prévia dos dados carregados:")
df.head()
```

 Análises Automatizadas (KPIs)

```python
#1- Vendas por produto
vendas_por_produto = df.groupby('Produto')['Valor'].sum()
print("Vendas por Produto:")
print(vendas_por_produto)
```

```python
#convertendo a coluna de data para tipo datetime
df['Data'] = pd.to_datetime(df['Data'])

#criando a coluna 'mes' a partir da coluna de data
df['mes'] = df['Data'].dt.to_period('M') # formato ano-mês
```

```python
#2- Vendas por mês
vendas_por_mes = df.groupby('mes')['Valor'].sum()
print("Vendas por Mes")
print(vendas_por_mes)
```

```python
#3- Vendas por vendedor
vendas_por_vendedor = df.groupby('Vendedor')['Valor'].sum()
print("Vendas por vendedor:")
print(vendas_por_vendedor)
```

```python
#4- ticket médio por vendedor
ticket_medio = df.groupby('Vendedor')['Valor'].mean()
print("Ticket médio por Vendedor:")
print(ticket_medio)

# média dos valores vendidos por cada vendedor = qualidade da venda
```

```python
#5- CRIAR O DASHBOARD (GRÁFICOS)

plt.style.use('ggplot') # deixa os gráficos com estilo mais bonito

# criar espaço de 2x2 gráficos (dashboard)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# gráfico 1- vendas por produto
vendas_por_produto.plot(kind='bar', ax=axes[0, 0],
title='Vendas Totais por produto', color='skyblue')

# gráfico 2- Vendas por Mês
vendas_por_mes.plot(kind='line', marker='o', ax=axes[0, 1],
title='Vendas por Mês', color='green')

# gráfico 3- vendas por vendedor (usando seaborn)
sns.barplot(x='Vendedor', y='Valor', data=df, ax=axes[1, 0],
palette='Blues')
axes[1, 0].set_title('Total de Produtos Vendidos por Vendedor')

# gráfico 4- ticket médio por vendedor
ticket_medio.plot(kind='barh', ax=axes[1, 1],
title='Ticket Médio por Vendedor', color='purple')

# Ajusta o layout automático
plt.tight_layout()
plt.show()
```
